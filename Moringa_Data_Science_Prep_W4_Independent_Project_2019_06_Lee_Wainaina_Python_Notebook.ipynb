```python
# Dropping irrelevant Data
# dropped the "Scheduled at" column cause it contains NULL values
df.drop(["Geo point", "Displayed comment", "Address", "Scheduled at"], axis = 1, inplace = True) 
```

```python
# checking for missing values
df.isnull().sum()
```

```python
# check for duplicates
df1 = df.drop_duplicates()
df1.head()
```

```python
# Outliers
# Data Cleaning Action: checking for outliers using IQR Scores
#


Q1 = df1.quantile(0.25)
Q3 = df1.quantile(0.75)

IQR = Q3 - Q1


(df1 < (Q1 - 1.5 * IQR)) |(df1> (Q3 + 1.5 * IQR))


# The two commented lines of code below can be used to drop the outliers but for now I will not drop them

#df1_no_outlier = dfc[~((df1 < (Q1 - 1.5 * IQR)) | (df1 > (Q3 + 1.5 * IQR))).any(axis = 1)]

# df1_no_outlier
```

```python
# sorting ID of the station
df1.sort_values(["ID","day","hour","minute"], ascending=[True,True,True,True])[10:50]
```

```python
# Hint:
# To compute usage, we will need to understand we have to join successive 
# (in time) measures/counters for a given station, as the difference will tell whether a car was picked up, returned, or nothing happened.
# Creating a new column where we find the difference in the number of bluecars at the station/counter

df1['car_Difference'] = df['Bluecar counter'].diff()
df1.head()
```

```python
# Q1

# Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018
df1[df1['car_Difference'] < 0].groupby('hour')['hour'].count().sort_values(ascending= False)
 

```

```python
# Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018

df1_paris = df1[df1['City'] =='Paris'] 

df1_paris.head()
```

```python
#  What postal code is the most popular for picking up Bluecars? Does the most popular station belong to that postal code?
 # Overall?

df1_paris[(df1_paris['Kind'] == 'STATION') & (df1_paris['Status'] == 'ok')].groupby('Postal code')['Postal code'].count().sort_values(ascending= False)
```

```python
# Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018

df1_paris[df1_paris['car_Difference'] < 0].groupby('hour')['hour'].count().sort_values(ascending= False)
```

```python
# What is the most popular hour for returning cars?

df1_paris[df1_paris['car_Difference'] > 0].groupby('hour')['hour'].count().sort_values(ascending= False)
```

```python
# What station is the most popular?
# Overall?

df1_paris[(df1_paris['Kind'] == 'STATION') & (df1_paris['Status'] == 'ok')].groupby('Public name')['Public name'].count().sort_values(ascending= False)
```
